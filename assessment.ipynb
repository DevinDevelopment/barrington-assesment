```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Importing database#
clothing_store = pd.read_csv("clothing_store_dataset (1).csv")
clothing_store
```

```python
#Replacing string values to null to be replaced#
clothing_store=clothing_store.replace('6N/A',np.nan).replace('3N/A',np.nan).replace('4N/A',np.nan)
clothing_store
```

```python
#Dropping cells with null values#
clothing_store=clothing_store.dropna()
clothing_store
```

```python
#Changing revenue column to conatin all float values#
clothing_store['revenue'].astype(float)
clothing_store
```

```python
#Confirming coloumns contain same datatypes#
clothing_store.dtypes
```

```python
#Confirming revenue count to further create visualizations#
clothing_store.loc[clothing_store['repeat_customer'] == 1, 'revenue'].sum()
```

```python
clothing_store.loc[clothing_store['repeat_customer'] == 0, 'revenue'].sum()
```

```python
#Visualizations#
plt.title('new(0) v. repeated(1) customers')
clothing_store.groupby('repeat_customer')['revenue'].sum().plot(kind='bar', ylim=(0,3500000))
plt.ylabel('Revenue(in millions)')
plt.show()
```

```python
#Confirming customer count based on new or returning condition#
new_cust=clothing_store.loc[clothing_store['repeat_customer'] == 0].count()[0]
new_cust
returned_cust=clothing_store.loc[clothing_store['repeat_customer'] == 1].count()[0]
returned_cust
```

```python
labels=['New Customers','Returned Customers']
plt.pie([new_cust,returned_cust], labels=labels, autopct='%.2f %%')
plt.title('percentage of new v. returned customers')
plt.show()
```

```python
plt.title('revenue of stores')
clothing_store.groupby('store')['revenue'].sum().plot(kind='bar', ylim=(0,1000000))
plt.ylabel('Revenue')
plt.show()
```
